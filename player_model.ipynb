```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
```

```python
#Load and parse data
df_match_data = pd.read_csv('/Users/sybren/Downloads/FIFA DataSet/Data/FIFA World Cup 2022 Match Data/data.csv', delimiter=',')
df_player_defense = pd.read_csv('/Users/sybren/Downloads/FIFA DataSet/Data/FIFA World Cup 2022 Player Data/player_defense.csv', delimiter=',')
df_match_data[['score_home', 'score_away']] = df_match_data.score.str.split("–", expand=True,)
```

```python
#Functions needed 
def PlayerRating(df):
    data = []
    #match = []
    #df_rating = pd.DataFrame()
    """Split the data per match
    """
    for i in df.index:
        print(i)
        team_home = df_player_defense.loc[df_player_defense['team'] == df['home_team'][i]].reset_index()
        team_away = df_player_defense.loc[df_player_defense['team'] == df['home_team'][i]].reset_index()
        match = (df['home_team'][i] + ' - ' + df['away_team'][i])
        print(match)
        for player in team_home.index:
            print(player)
            if team_home['position'][player] == 'DF':
                DefenderRating(df, i)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : DefenderRating.score_def[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_home['position'][player] == 'MF':
                MidfielderRating(df, i)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : MidfielderRating.score_mid[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_home['position'][player] == 'FW':
                AttackerRating(df, i )
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : AttackerRating.score_att[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)

        global df_rating 
        df_new = pd.DataFrame(data)
        df_rating = pd.concat([df_rating, df_new])
        #df_rating.set_index(['match'])
        #df_new = pd.DataFrame(data)
        #df_rating = pd.concat([df_rating, df_new])
    return df_rating

def DefenderRating(df, i):
    """Calculate individual score of each defender in a match
    """
    if df['score_home'][i] == 0:
        clean_sheet = 5
    else:
        clean_sheet = 0 
    try: 
        DefenderRating.score_def = clean_sheet - (int(df['score_away'][i])*(1/(df['away_xg']))) + df['home_clearances'][i]
        print(DefenderRating.score_def)
    except:
        pass;

def MidfielderRating(df, i):
    """Calculate individual score of each midfielder in a match
    """
    try: 
        MidfielderRating.score_mid =  -0.5*(int(df['score_home'][i])*(1/(df['home_xg']))) + df['home_interceptions'][i] + df['home_tackles']
        #print(DefenderRating.score_def)
    except:
        pass;

def AttackerRating(df, i):
    """Calculate individual score of each midfielder in a match
    """
    try: 
        AttackerRating.score_att =  3*(int(df['score_home'][i])*(1/(df['home_xg']))) + df['home_sot']
        #print(DefenderRating.score_def)
    except:
        pass;

```

```python
#Call function
df_rating = pd.DataFrame()
PlayerRating(df_match_data)
```
