```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
```

```python
#Load and parse data
df_match_data = pd.read_csv('../FIFA DataSet/Data/FIFA World Cup 2022 Match Data/data.csv', delimiter=',')
df_player_defense = pd.read_csv('../FIFA DataSet/Data/FIFA World Cup 2022 Player Data/player_defense.csv', delimiter=',')
df_match_data[['score_home', 'score_away']] = df_match_data.score.str.split("–", expand=True,)
```

```python
#Functions needed 
def PlayerRating(df):
    data = []
    #match = []
    #df_rating = pd.DataFrame()
    """Split the data per match
    """
    for i in df.index:
        print(i)
        team_home = df_player_defense.loc[df_player_defense['team'] == df['home_team'][i]].reset_index()
        team_away = df_player_defense.loc[df_player_defense['team'] == df['home_team'][i]].reset_index()
        match = (df['home_team'][i] + ' - ' + df['away_team'][i])
        print(match)
        for player in team_home.index:
            #print(len(team_home), len(team_away))
            if team_home['position'][player] == 'DF':
                team = 'home'
                DefenderRating(df, i, team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : DefenderRating.score_def_home[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_home['position'][player] == 'MF':
                team = 'home'
                MidfielderRating(df, i, team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : MidfielderRating.score_mid_home[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_home['position'][player] == 'FW':
                team = 'home'
                AttackerRating(df, i , team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : AttackerRating.score_att_home[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)

        for player in team_away.index:
            #print(len(team_home), len(team_away))
            if team_away['position'][player] == 'DF':
                team = 'away'
                DefenderRating(df, i, team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : DefenderRating.score_def_away[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_away['position'][player] == 'MF':
                team = 'away'
                MidfielderRating(df, i, team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : MidfielderRating.score_mid_away[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)
            if team_away['position'][player] == 'FW':
                team = 'away'
                AttackerRating(df, i , team)
                #df_rating[['name', 'rating']] = (team_home['player'][i], DefenderRating.score_def)
                #df_rating['name'][player] = team_home['player'][player]
                #df_rating['rating'][player] = DefenderRating.score_def
                new = {'name' : team_home['player'][player], 'rating' : AttackerRating.score_att_away[0], 'match': match}
                print(new)
                data.append(new)
                #data.append(new)

        global df_rating 
        df_new = pd.DataFrame(data)
        df_rating = pd.concat([df_rating, df_new])
        #df_rating.set_index(['match'])
        #df_new = pd.DataFrame(data)
        #df_rating = pd.concat([df_rating, df_new])
    return df_rating

def DefenderRating(df, i, team):
    """Calculate individual score of each defender in a match
    """
    if df['score_away'][i] == 0:
        clean_sheet_home = 5
    else:
        clean_sheet_home = 0 
    if df['score_home'][i] == 0:
        clean_sheet_away = 5
    else:
        clean_sheet_away = 0 

    if team == 'home':
    
        try: 
            DefenderRating.score_def_home = clean_sheet_home - (int(df['score_away'][i])*(1/(df['away_xg']))) + df['home_clearances'][i]
            print(DefenderRating.score_def)
        except:
            pass;
    if team == 'away':
        try: 
            DefenderRating.score_def_away = clean_sheet_away - (int(df['score_home'][i])*(1/(df['home_xg']))) + df['away_clearances'][i]
            print(DefenderRating.score_def)
        except:
            pass;
 

def MidfielderRating(df, i, team):
    """Calculate individual score of each midfielder in a match
    """
    if team == 'home':
        try: 
            MidfielderRating.score_mid_home =  -0.5*(int(df['score_away'][i])*(1/(df['away_xg']))) + df['home_interceptions'][i] + df['home_tackles']
            #print(DefenderRating.score_def)
        except:
            pass;
    if team == 'away':
        try: 
            MidfielderRating.score_mid_away =  -0.5*(int(df['score_home'][i])*(1/(df['home_xg']))) + df['away_interceptions'][i] + df['away_tackles']
            #print(DefenderRating.score_def)
        except:
            pass;

def AttackerRating(df, i, team):
    """Calculate individual score of each midfielder in a match
    """
    if team == 'home':
        try: 
            AttackerRating.score_att_home =  3*(int(df['score_home'][i])*(1/(df['home_xg']))) + df['home_sot']
            #print(DefenderRating.score_def)
        except:
            pass;
    if team == 'away':
        try: 
            AttackerRating.score_att_away =  3*(int(df['score_home'][i])*(1/(df['home_xg']))) + df['home_sot']
            #print(DefenderRating.score_def)
        except:
            pass;

```

```python
#Call function
df_rating = pd.DataFrame()
PlayerRating(df_match_data)
```
